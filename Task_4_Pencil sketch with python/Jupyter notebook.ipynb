```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Read the image in RGB format

```python
image_path = 'puppy.jpg'  # Replace 'path_to_image.jpg' with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
```

## Display the original image

```python
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Convert the image to grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Invert the grayscale image (negative image)

```python
inverted_image = cv2.bitwise_not(gray_image)
```

## Blur the inverted image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0)
```

```python
# Step 6: Adjust the intensity of the blurred image
adjusted_blurred = cv2.addWeighted(blurred_image, 1.5, np.zeros_like(blurred_image), 0, 0)
```

## Create the pencil sketch by dividing the grayscale image by the inverted blurred image

```python
pencil_sketch = cv2.divide(gray_image, 255 - adjusted_blurred, scale=256.0)
plt.figure(figsize=(8, 8))
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```

## Save the pencil sketch image

```python
output_path = 'pencil_sketch.jpg'
_ = cv2.imwrite(output_path, pencil_sketch)
```
